# Libraries

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator
import matplotlib.image as mpimg
from scipy.fft import fft2, ifft2
```

## Read image

```python
img = mpimg.imread('Images/cameraman.tif')
```

## Plotting image

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## DFT

```python
img_DFT = fft2(img)
```

## Set DC term to zero

```python
img_DFT[0][0] = 0.0
```

## Inverse DFT

```python
img_iDFT = abs(ifft2(img_DFT))
```

## Plotting image rewrite with DC set to zero

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img_iDFT, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## Average of all pixels

```python
avg = img.mean()
```

## Subtract average from image

```python
img_processed = abs((img - avg))
```

## Plotting image processed

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img_processed, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## Plotting both images

```python
fig, axs = plt.subplots(1,2, figsize = (18,7))

axs[0].imshow(img_iDFT, cmap = 'gray')
axs[1].imshow(img_processed, cmap = 'gray')

titles = ['(a)\n', '(b)\n']

for ax, title in zip(axs, titles):
        
    # Set title
        
    ax.set_title(title, fontsize = 24)
        
    # Remove axis
        
    ax.xaxis.set_major_locator(NullLocator())
    ax.yaxis.set_major_locator(NullLocator())
    
plt.savefig("Results/F0.pdf", bbox_inches = 'tight')
```
