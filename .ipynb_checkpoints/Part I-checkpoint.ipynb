## Libraries

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.ticker import NullLocator
import matplotlib.image as mpimg
```

## Read image

```python
img = mpimg.imread('Images/cameraman.tif')
```

## Plotting image

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## DFT

```python
from scipy.fft import fft2, ifft2
```

```python
img_DFT = fft2(img)
```

```python
img_DFT_real, img_DFT_imag = img_DFT.real, 1.0j*img_DFT.imag
```

## DFT reverse

```python
img_real = ifft2(img_DFT_real)
```

```python
img_imag = ifft2(img_DFT_imag)
```

## Plotting images

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img_real.real, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img_imag.real, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```
