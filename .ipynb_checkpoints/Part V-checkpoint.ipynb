## Libraries

```python
import numpy as np
import cmath
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import NullLocator
import matplotlib.image as mpimg
from scipy.fft import fft2, ifft2
```

## Read images

```python
img = mpimg.imread('Images/cameraman.tif')
```

## Plotting image

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## DFT

```python
img_DFT = fft2(img)
```

## Get the amplitude

## $F(u,v)=|F(u,v)|$

```python
img_DFT_amplitude = abs(img_DFT)
```

## Inverse DFT of amplitude phase of the frenquency spectrum

```python
img_amplitude = np.sqrt(abs(ifft2(img_DFT_amplitude)))
```

## Plotting image rewrite from amplitude of the frenquency spectrum

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img_amplitude, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## Get the phase of DFT

```python
M, N = img_DFT.shape

img_DFT_phase = np.zeros_like(img_DFT)

for m in range(M):
    for n in range(N):
        img_DFT_phase[m][n] = cmath.phase(img_DFT[m][n])
        
img_DFT_phase = np.exp(img_DFT_phase*1j)
```

## Inverse of DFT phase

```python
img_phase = abs(ifft2(img_DFT_phase))
```

## Plotting image rewrite from phase of frequency spectrum

```python
fig, ax = plt.subplots(figsize = (8,6))

ax.imshow(img_phase, cmap = 'gray')

plt.gca().xaxis.set_major_locator(NullLocator())
plt.gca().yaxis.set_major_locator(NullLocator())
```

## Plotting both images

```python
fig, axs = plt.subplots(1,2, figsize = (18,7))

axs[0].imshow(img_amplitude, cmap = 'gray')
axs[1].imshow(img_phase, cmap = 'gray')

titles = ['(a)\n', '(b)\n']

for ax, title in zip(axs, titles):
        
    # Set title
        
    ax.set_title(title, fontsize = 24)
        
    # Remove axis
        
    ax.xaxis.set_major_locator(NullLocator())
    ax.yaxis.set_major_locator(NullLocator())
    
plt.savefig("Results/amplitude_phase.pdf", bbox_inches = 'tight')
```
